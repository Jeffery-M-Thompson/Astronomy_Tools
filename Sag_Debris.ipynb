### Plotting the Selected Data

The data was plotted in polar plots of the south Galactic Cap in galactic coordinates $(l, b)$. The stars were arranged into bins of equal area. The process began by dividing circle around the pole of unit radius into $3$ parts. The area of one of these parts was used to find the number of divisions for each additonal unit radius out from the center. If we start with radius $r_0 = 0$ and our first circle is $r_1=1$, divided into three equal parts, each part covering and angle of $\theta_0=\frac{2 \pi}{3}$. So the area of each polar segment is $a=\frac{\pi r_1^2}{3}$. So each radius increase of 1 unit needs to be divided into segments that equal this in area.

Which can be expressed by the following relationship. If the distance in $r$ are one unit.
$$n_i = 3(i^2-(i-1)^2)\\
=3(2i-1)\\
=6i-3\\
\theta_i=\frac{2\pi}{n_i}\\
bins_i=6\sum\limits_{k=1}^i k - \sum\limits_{k=1}^i 3= 3i(i+1)-3i=3i^2$$

This maps out a polor plot into bins as follows an example for $432$ bins:
</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from random import choice

# bin properties
radius = 12
no_r_bins = 12                      # number of r sized bins in the radial direction
delta_r = radius/float(no_r_bins)   # we use a unit size for the bin
no_bins = 3*no_r_bins**2            # total number of bins
bins =np.zeros((no_bins, 5))        # Generate bins 

# create bin boundaries
fig = plt.figure(num=None, figsize=(15, 15), dpi=300, facecolor='w', edgecolor='k')
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)

bin_no = 0
for row in range(no_r_bins):
    n_i = 6*(row+1)-3
    delta_theta = (2.0*np.pi)/float(n_i)
    r_min = row * delta_r
    r_max = (row+1) * delta_r
    theta_min = 0.0
    for cell in range(n_i):
        theta_max = theta_min + delta_theta
        bins[bin_no]=[r_min, r_max, theta_min, theta_max, bin_no]
        theta = np.arange(theta_min, theta_max, 1.0/180)
        r_1 = r_min + 0*theta
        r_2 = r_max + 0*theta
        plt.polar(theta, r_1, lw=1, c = 'red')
        plt.polar(theta, r_2, lw=1, c = 'red')
        plt.fill_between(theta, r_1, r_2, alpha=0.45, edgecolor = 'red', interpolate = True)
        ax.annotate(str(bin_no),
            xy=(
                ((theta_max+theta_min)/2.0),
                ((r_max+r_min)/2.0)),
            xytext=(
                ((theta_max+theta_min)/2.0),
                ((r_max+r_min)/2.0)),
            textcoords="data",
            color='yellow',
            fontsize =10)
        theta_min = theta_max
        bin_no = bin_no + 1

plt.ylim(0,radius)
ax.set_yticks([])
plt.show()
```
